```python
import os
import pandas as pd
import numpy as np
```

```python
# Path to the .csv files with articles' information
DF_input = input().strip()
```

```python
# Path to the file name with prepared institutions
DF_output = input().strip()
```

```python
# Import list of csvs to parse
list_csvs = []

for file in os.listdir(DF_input):
    if file[-4:] == ".csv":
        list_csvs.append(file)

n_csvs = len(list_csvs)
```

```python
list_rows = []

for csv in list_csvs[0:1] + list_csvs[30:]:
    print(f"Parsing csv {csv}")
    df = pd.read_csv(DF_input + csv)
    df = df.loc[df["Authors"].notna()]
    df = df.iloc[2:4]
    list_rows.append(df)
    del df
```

```python
df = pd.concat(list_rows)
df = df.reset_index()[["PMID", "Authors", "NER_BERT", "Country"]]
```

```python
df["Institutions_man"] = None
df["Institutions_man"] = df["Institutions_man"].astype("object")
df["Countries_man"] = None
df["Countries_man"] = df["Countries_man"].astype("object")
```
